## GMX Backend Queries

```python
import json
import pandas as pd
from websocket import create_connection
```

```python
ws = create_connection("wss://www.gmx.house/api-ws")
```

```python
# query leaderboard
ws.send('{"topic":"requestLeaderboardTopList","body":{"timeInterval":16934400,"offset":0,"pageSize":10000,"sortBy":"realisedPnl","chain":42161,"sortDirection":"desc"}}')
```

```python
result =  ws.recv()
```

```python
data = json.loads(result)
```

```python
leaderboard = pd.DataFrame(data['body']['page']).drop(columns=['claim'])
```

```python
leaderboard
```

```python
leaderboard.to_csv('all_time_leaderboard.csv')
```

```python
START_TIME = 16934400  # all time
```

```python
def get_account_request(account, time_interval):
    return '{"topic":"requestAccountTradeList","body":{"account":"' + str(
        account) + '","timeInterval":' + str(time_interval) + ',"chain":42161}}'
```

```python
accounts = leaderboard['account'].to_list()
```

```python
dfs = []
```

```python
for account in accounts:
    ws.send(get_account_request(account, START_TIME))
    result =  ws.recv()
    data = json.loads(result)
    dfs.append(pd.DataFrame(data['body']))
```

```python
trade_data = pd.concat(dfs).reset_index(drop=True)
```

```python
trade_data.to_csv('gmx_trade_data.csv')
```
